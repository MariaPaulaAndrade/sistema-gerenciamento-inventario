## Objetivo <a id='objetivo'></a>

O principal objetivo deste projeto  desenvolver uma soluo de gerenciamento de inventrio de produtos para uma loja virtual que:
* Permita o cadastro, atualizao, excluso e listagem de produtos.
* Contemple a variedade de produtos vendidos pela loja que variam em categorias, preos e quantidade disponveis.
* Permita o gerenciamento de produtos, facilitar operaes de vendas, armazenamento de dados e gerao de relatrios detalhados em CSV.
* Seja robusto (escalvel), eficiente (legvel e menos custoso computacionalmente) e capaz de lidar com erros de forma elegante.
* Oferea menus e opes que guiem usurio durante o uso do sistema de forma flexvel e permitindo a customizao das funcionalidades.
* Oferea um feedback claro ao usurio em caso de erros e garanta que o sistema continue funcionando corretamente.
* Interaja com o usurio de forma amigvel, clara e responsiva.

### Interface com o usurio <a id='interface'></a>

#### Funo de Interao com o Usurio

```python
user_interface(comando, path_produtos, path_vendas)
```
- **Descrio:** Permite que o usurio interaja com um sistema de gerenciamento de inventrio. Oferece as opes de visualizar, cadastrar, atualizar e excluir produtos, visualizar relatrio de vendas e efetu-las. Carrega e salva informaes de produtos e vendas em arquivos CSV. No retorna valores, apenas executa comandos e manipula arquivos de produtos e vendas.
- **Parmetros:**
  - `comando` (int): Um nmero inteiro que indica a opo desejada pelo usurio. Os valores e opes so:
    - `1`: Visualizar produtos.
    - `2`: Cadastrar um novo produto.
    - `3`: Atualizar um produto existente.
    - `4`: Excluir um produto.
    - `5`: Visualizar  o relatrio de vendas.
    - `6`: Efetuar uma venda.
  - `path_produtos` (str): Caminho para o arquivo CSV que contm a lista de produtos.
  - `path_vendas` (str): Caminho para o arquivo CSV que contm a lista de vendas.
- **Funcionamento:**
  1. **Carregamento de Dados:**
     - Carrega lista de produtos a partir do arquivo CSV especificado em `path_produtos`.
     - Carrega lista de vendas a partir do arquivo CSV especificado em `path_vendas`.
  2. **Execuo de Comandos:**
     - **Comando 1:** Visualiza a lista de produtos ativos usando a funo `lista_produtos_ativos`.
     - **Comando 2:** Cadastra um novo produto com os dados recebidos do usurio. O produto  adicionado a lista com um identificador nico (ID incrementado).
     - **Comando 3:** Atualiza um produto existente com base no ID do produto e nas informaes que o usurio desejar alterar.
     - **Comando 4:** Exclui um produto da lista com base no ID fornecido pelo usurio. O produto fica inativado no sistema.
     - **Comando 5:** Visualiza o relatrio de vendas utilizando a funo `lista_vendas`.
     - **Comando 6:** Inicia uma nova venda.
  3. **Persistncia dos Dados:**
     - Aps a execuo de comandos, a funo de interface do usurio salva as listas atualizadas de produtos e vendas nos arquivos CSV especificados.
  4. **Funes de suporte a interface do usurio:**
     - `carrega_produtos_csv()`
     - `carrega_vendas_csv()`
     - `lista_produtos_ativos()`
     - `cadastra_produto()`
     - `atualiza_produto()`
     - `desativa_produto()`
     - `lista_vendas()`
     - `inicia_venda()`
     - `persistir_produtos_csv()`
     - `persistir_vendas_csv()`
  5. **Observaes:**
     - Os tratamentos de exceo so feitos dentro das funes de suporte.
     

#### Funo de Gerenciamento de Fluxo de Execuo

```python
main()
```
- **Descrio:** Inicia e gerencia o fluxo de execuo do sistema de gerenciamento.  o ponto de partida principal do programa.
- **Funcionamento:**
  1. Obtm o diretrio do trabalho e cria um caminho para uma pasta chamada 'dados' dentro do diretrio.
  2. Verifica se a pasta 'dados' j existe. Caso contrrio, ela  criada.
  3. Define caminho para dois arquivos CSV, 'produtos.csv' e 'vendas.csv', que vo estar localizados na pasta 'dados'.
  4. Entra em loop exibindo um menu e solicitando ao usurio que escolha uma opo.
  5. Dependendo da opo escolhida pelo usurio, a funo `user_interface`  chamada com os caminhos dos arquivos como argumentos. Se a funo retorna `True` o loop  interrompido e exibe uma mensagem de agtradecimento ao usurio.
  6. O programa continua em execuo at que o usurio decida finalizar. O programa exibe uma mensagem de agradecimento e encerra.
- **Observaes:**
  - O menu e as opes disponveis so gerenciadas pela funo principal `menu` e a entrada do usurio  processada pela funo auxiliar `input_numero()`.

#### Funo de Cadastro de Produtos

```python
cadastra_produto(produtos, produto_id, produto_nome, produto_descricao, produto_quantidade, produto_valor, produto_categoria)
```

- **Descrio:** Permite o registro de novos produtos no sistema de gerenciamento e captura informaes de nome, categoria, preo, quantidade em estoque e descrio opcional do produto.
- **Atributos:**
    - Recebe os dados de entrada do usurio.
    - Armazena os dados em uma estrutura de dicionrio onde cada produto recebe um identificador nico.
    - Classifica os produtos como ativos no registro.
    - Retorna ao usurio confirmao de registro do novo produto.
- **Parmetros**
    - `produtos` (list): A lista de produtos onde o novo produto ser adicionado. Cada produto na lista  representado por um dicionrio.
    - `produto_id` (int): O identificador nico do produto.
    - `produto_nome` (str): O nome do produto.
    - `produto_descricao` (str): A descrio do produto.
    - `produto_quantidade` (int): A quantidade disponvel do produto em estoque.
    - `produto_valor` (float): O preo do produto.
    - `produto_categoria` (str): A categoria do produto.

#### Funo de Atualizao de Produtos
```python
atualiza_produto(produtos, produto_id, **produto_atualizacao)
```
- **Descrio:** Atualiza informaes de produtos cadastrados no sistema.
- **Atributos:**
    - Recebe o identificador nico do produto cadastrado.
    - Recebe o tipo de informao que o usurio deseja atualizar (nome, categoria, preo, estoque ou descrio).
    - Armazena a informao atualizada do produto.
    - Retorna ao usurio mensagem de atualizao cadastral e aviso caso o produto no tenha sido encontrado.
- **Parmetros:**
    - `produtos` (list): A lista de produtos, onde cada produto  representado por um dicionrio. O dicionrio deve conter pelo menos a chave 'id'.
    - `produto_id` (int): O identificador nico do produto a ser atualizado. A funo procura o produto com esse ID na lista de produtos.
    - `**produto_atualizacao`: Um nmero varivel de argumentos de palavra-chave que representam as atualizaes a serem feitas. As chaves devem corresponder s chaves existentes no dicionrio do produto e podem incluir:
        - 'nome' (str): O novo nome do produto.
        - 'descricao' (str): A nova descrio do produto.
        - 'quantidade' (int): A nova quantidade disponvel do produto.
        - 'valor' (float): O novo valor do produto. Se fornecido como string, ser convertido para float.
        - 'categoria' (str): A nova categoria do produto.
        - 'ativo' (bool): Indica se o produto est ativo ou no. Se a chave 'ativo' for fornecida como False, o produto ser marcado como inativo.

#### Funo de Listagem de Produtos Ativos
```python
lista_produtos_ativos(produtos)
```
- **Descrio:** Lista os produtos que esto ativos para serem vendidos, manipulados, atualizados e includos em relatrios pelo gerenciador do sistema.

#### Funo de Inativao (Excluso) de Produtos

- **Atributos:**

    - Retorna mensagem de que o produto foi excludo do sistema ou que no foi encontrado no sistema.
- **Parmetros:**
    - `produto_id` (int): O identificador nico do produto a ser desativado. A funo busca o produto na lista com este ID.
    - `produtos` (list): Uma lista de dicionrios onde cada dicionrio representa um produto. Cada produto deve conter pelo menos as chaves 'id' e 'ativo'. O valor da chave 'ativo' indica se o produto est ativo (True) ou inativo (False).

#### Funo de Gerao de Recibos de Venda
```python
gera_recibo(venda)
```
- **Descrio:** Gera um recibo de venda em formato CSV e o salva em um arquivo.
- **Atributos:**
  - Recebe um dicionrio com os dados da venda.
  - Cria um arquivo CSV contendo os detalhes de uma venda, incluindo o ID da venda, a lista de produtos, o valor total e a data e hora da venda.
  - Salva o arquivo em uma pasta chamada 'recibos' no diretrio de trabalho atual. Se a pasta no existir, ela ser criada.
- **Parmetros:**
    - `venda` (dict): Um dicionrio contendo as informaes da venda. Deve incluir as seguintes chaves:
        - 'id' (int): O identificador nico da venda.
        - 'produtos' (list): Lista de produtos envolvidos na venda, onde cada produto  representado como um dicionrio.
        - 'valor_total' (float): O valor total da venda.
        - 'data_hora' (str): A data e hora em que a venda foi realizada, no formato 'YYYY-MM-DD HH:MM:SS'. 

#### Funo de Atualizao de Estoque
```python
atualiza_estoque(produtos, produtos_venda)
```
- **Descrio:** Atualiza o estoque de produtos, carregado e inicializado no comeo do programa.
- **Atributos:**

- **Parmetros:**
  - `produtos` (list of dict): Lista de produtos, onde cada produto  representado como um dicionrio com as chaves 'id' (identificador do produto) e 'quantidade' (estoque atual).
  - `produtos_venda` (list of dict): Lista de produtos vendidos, onde cada produto vendido  representado como um dicionrio com as chaves 'id' (identificador do produto) e 'quantidade' (quantidade vendida).

#### Funo de Finalizao de Vendas
```python
finaliza_venda(produtos, vendas, *venda_produtos)
```
- **Descrio:** Finaliza uma venda, calcula o valor total, registra a venda e gera um recibo.

  - `vendas` (list): Uma lista onde cada item  um dicionrio representando uma venda registrada. A funo adicionar um novo dicionrio de venda a esta lista.
  - `*venda_produtos`: Um nmero varivel de argumentos, onde cada argumento  um dicionrio representando um produto vendido na transao. Cada dicionrio deve conter as chaves 'id', 'nome', 'valor', e 'quantidade'.

  - `*produtos_vendas` (dict): Lista de produtos vendidos, onde cada produto vendido  representadocomo um dicionrio com as chaves 'id' (identificador do produto) e'quantidade' (quantidade vendida).

O dicionrio pode conter outras chaves como 'nome' (nome do produto), mas estas no so obrigatrias para a funo funcionar.

#### Funo de Tratamento de Dados Numricos

```python
input_numero(prompt, opcional=False, tipo: Type = int, valores_validos=None, tam_min=0, tam_max=9999999)
```
- **Descrio:** Solicita, valida a entrada do usurio e a converte para o tipo especificado.

  - `valores_validos` (list of int, opcional): Lista de valores vlidos que a entrada deve corresponder. Se `None`, qualquer valor  aceito.
  - `tam_min` (int, opcional): O valor mnimo de caracteres permitido. O padro  0.
  - `tam_max` (int, opcional): O valor mximo de caracteres permitido. O padro  9999999.

#### Funo de Carregamento de Vendas
```python
carrega_vendas_csv(path)
```
- **Descrio:** Carrega uma lista de vendas a partir de um CSV. 
- **Atributos:**

#### Funo de Persistncia de Vendas
```python
persistir_vendas_csv(vendas, path)
```
- **Descrio:** Salva uma lista de vendas em um arquivo CSV.
- **Atributos:**
  - Escreve uma lista de vendas em um arquivo CSV. Cada venda deve ser representada como um dicionrio com chaves correspondentes aos nomes das colunas no arquivo CSV.
  - Converte a coluna 'produtos' de uma lista de dicionrios para uma string JSON antes de escrever.
  - Exibe mensagens de erro se o arquivo no puder ser acessado ou se ocorrer um erro ao processar o arquivo ou seus dados.
- **Parmetros**
  - `vendas` (list of dict): Lista de dicionrios, onde cada dicionrio representa uma venda com as seguintes chaves:
    - 'id': ID da venda (int)
    - 'produtos': Lista de produtos vendidos. Cada produto  representado como um dicionrio
    - 'valor_total': Valor total da venda (float)
    - 'data_hora': Data e hora da venda (str)
  - `path` (str): O caminho para o arquivo CSV onde as vendas sero salvas.

### Classificao das funes <a id='classificacao'></a>

1. **Criao, leitura, atualizao e excluso (CRUD)**
- **Propsito:** Adiciona, recupera, visualiza, modifica e remove registros armazenados.
- **Funes:**
  - Cadastrar produtos -> `cadastra_produto()`
  - Atualizar produtos -> `atualiza_produto()`
  - Listar produtos ativos -> `lista_produtos_ativos()`
  - Inativar (excluir) produtos -> `desativa_produto()`

2. **Vendas**
- **Propsito:** Operacionaliza todo o processo de venda do sistema de gerenciamento.
- **Funes:**
  - Elaborar recibo de vendas -> `gera_recibo()`
  - Atualizar estoque -> `atualiza_estoque()`
  - Finalizar uma venda -> `finaliza_venda()`
  - Processar uma venda -> `gera_venda()`
  - Verificar a quantidade do produto em estoque -> `valida_produto()`
  - Iniciar o processo de venda -> `inicia_venda()`
  - Elaborar relatrio de vendas -> `lista_vendas()`

3. **Suporte e Tratamento de Excees**
- **Propsito:** Auxilia e garante o correto funcionamento do cdigo evitando casos com o cdigo repetitivo, tratando e visualizando possveis erros (excees).
- **Funes:**
  - Converter dados nmericos no padro numrico (com ponto em caso de nmeros decimais)  -> `converte_numero()`
  - Retornar mensagens de erros em campos que devem receber dados numricos -> `input_numero()`
  - Retornar mensagens de erros em campos que devem receber dados textuais -> `input_texto()`

4. **Armazenamento de dados e Elaborao de arquivos**
- **Propsito:** Garante a persistncia dos dados do inventrio e informaes de vendas.
- **Funes:**
  - Abrir e ler um arquivo (do tipo CSV) existente com os dados dos produtos -> `carrega_produtos_csv()`
  - Armazenar os dados do inventrio e as informaes de venda em arquivo CSV -> `persistir_arquivos_csv()`
  - Abrir e ler um arquivo (do tipo CSV) existente com os dados de venda -> `carrega_vendas_csv()`
  - Grava os dados de vendas em um arquivo CSV especificado pelo caminho fornecido -> `persistir_vendas_csv()`
 
5. **Interao e gerenciamento do sistema**
- **Propsito:** Permite a interao com o usurio e o gerenciamento do inventrio de produtos.
- **Funes:**
  - Oferecer ao usurio opes de gerenciamento do sistema -> `user_interface()`
  - Assegurar o fluxo de execuo e gerenciamento do sistema -> `main()`

### Decises de design <a id='decisoes'></a>

* Decidimos criar mais funes do que o requisitado para a elaborao do sistema para garantir a mudalaridade, facilitar a manipulao pelo usurio e assegurar que os mdulos (funes) podem ser atualizados e ajustados de forma independente no sistema. 
* Diferentemente da ao de excluir decidimos criar a funo de inativao do produto no sistema a partir de um "selo" de produto ativo (chave 'ativo'). Isso garante que a informao sobre o inventrio seja passada com maior controle e preciso no apenas sobre os produtos que esto disponveis como aqueles que estiveram.
